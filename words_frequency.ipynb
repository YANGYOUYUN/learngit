```python
import numpy as np
import pandas as pd
import jieba
import os
```

```python
comments = pd.read_table('./cut_words.txt',encoding='utf-8')
comments.rename(columns={'数据包':'title'},inplace=True)  #修改列名
comments.head()
```

```python
# 去停用词
stopwords = [line.rstrip() for line in open('./Stopword.txt',encoding='utf-8')]
# stopwords
```

```python
#自定义词典
user_dict = ['高活跃','高付费','高质量','高消费','高价值','高互动']
for i in user_dict:
    jieba.add_word(i)
```

```python
#分词
result = []
# jieba.add_word("高活跃")
with open('./cut_words.txt',encoding='utf-8') as f:
    for word in f.readlines():
        seg = jieba.lcut(word)
        for i in seg:
            if i not in stopwords and i not in ('_','\n'):
                result.append(i)
    print(result) 
    
    
```

```python
#统计词频
word_list= {}
for item in result:
    if item not in word_list:
        word_list[item]=1
    else:
        word_list[item]+=1
print(word_list)
```

```python
#对结果进行排序
key = list(word_list.keys())
value = list(word_list.values())
df = pd.DataFrame()
df['key'] =key
df['value'] = value
df.sort_values(by=['value'],ascending=False)[:60]
```
